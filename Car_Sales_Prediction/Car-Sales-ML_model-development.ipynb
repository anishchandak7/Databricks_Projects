# ML Model Development

## 1. Read data

```
# File location and type
file_location = "/FileStore/tables/formatted_car_sales_data.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# What is the use of inferSchema?
# inferSchema -> Infer schema will automatically guess the data types for each field.

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

## 2. Aim

### Build a machine learning model using car sales data to predict the selling price of a car.

## 3. Data Preparation

```
# Pyspark dataframe --> Pandas dataframe
df2 = df.toPandas()
```

```
# Remove unnecessary columns
redundant_cols = ['Date', 'Salesperson', 'Customer Name']
df2.drop(columns = redundant_cols, inplace = True)
```

```
df2.head()
```

### 1. Dtype Correction

```
df2.info()
```

```
df2['Car Year'] = df2['Car Year'].astype('int')
df2['Sale Price'] = df2['Sale Price'].astype('float')
df2['Commission Rate'] = df2['Commission Rate'].astype('float')
df2['Commission Earned'] = df2['Commission Earned'].astype('float')
df2['Year'] = df2['Year'].astype('int')
df2['Month'] = df2['Month'].astype('int')
```

```
df2.info()
```

### 2. Feature Engineering

```
df2['Car Age'] = df2['Year'] - df2['Car Year']
```

```
df2['Car Age'].unique()
```

```
df2.head()
```

```
df3 = df2.drop(columns = ['Car Year', 'Year', 'Month'])
```

```
df3.head()
```

```
Y = df3['Sale Price']
X = df3.drop(columns = ['Sale Price'])
```

```
X.head()
```

```
Y
```

### 3. Handling categorical features

```
cat_features = ['Car Make', 'Car Model']
for col in cat_features:
    X[col] = X[col].astype('category').cat.codes
```

```
X.head()
```

```
X.info()
```

## 4. Model Declaration

```
from sklearn import linear_model
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.tree import DecisionTreeRegressor
```

```
models = {
    'Ridge Regressor':linear_model.Ridge(),
    'Lasso Regressor':linear_model.Lasso(),
    'Elastic-Net':linear_model.ElasticNet(),
    'Linear Regression':linear_model.LinearRegression(),
#     'Logistic Regression':linear_model.LogisticRegression(),
    'Bayesian Regression':linear_model.BayesianRidge(),
#     'SGDRegressor':linear_model.SGDRegressor(),
#     'SVR':SVR(),
    'KNN Regressor': KNeighborsRegressor(),
#     'Gaussian Process Regressor': GaussianProcessRegressor(),
    'DecisionTree Regressor': DecisionTreeRegressor()
}
```

## 5. Split data

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y,shuffle=True)
```

## 6. Evaluation Metrics

```
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

## 7. Model Training & Evaluation

```
import seaborn as sns
import matplotlib.pyplot as plt
model_list = []
r2_list =[]
for model_name in models:
    
    # Pick a model
    model = models[model_name]
    
    # Fit the train dataset
    model.fit(X_train, y_train)
    
    # Predict on test dataset
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Evaluate model performance on train & test dataset.
    model_train_mae , model_train_rmse, model_train_r2 = evaluate_model(y_train, y_train_pred)
    model_test_mae , model_test_rmse, model_test_r2 = evaluate_model(y_test, y_test_pred)
    
    print(model_name)
    model_list.append(model_name)

    print('Model performance for Training set')
    print("- Root Mean Squared Error: {:.4f}".format(model_train_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_train_mae))
    print("- R2 Score: {:.4f}".format(model_train_r2))

    print('----------------------------------')
    
    print('Model performance for Test set')
    print("- Root Mean Squared Error: {:.4f}".format(model_test_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_test_mae))
    print("- R2 Score: {:.4f}".format(model_test_r2))
    r2_list.append(model_test_r2)
    
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.regplot(x=y_train, y=y_train_pred, ci=None, color='blue')
    plt.title('Regression plot for training dataset for model '+model_name)
    
    plt.subplot(1,2,2)
    sns.regplot(x=y_test, y=y_test_pred, ci=None, color='red')
    plt.title('Regression plot for testing dataset for model '+model_name)
    plt.show()

    print('='*35)
    print('\n')
    del model
```

```
import pandas as pd
results_df = pd.DataFrame(list(zip(model_list, r2_list)), columns=['Model Name', 'R2_Score']).sort_values(by=["R2_Score"],ascending=False)
```

```
results_df
```

### Observation:

* Decision Tree performs well with the given dataset with 99% accuracy. It's way to good!!
* All linear models (Bayesian, Linear, Ridge, Lasso) also works performed well with 92% accuracy.

## 8. Ensemble Models

```
from sklearn.ensemble import VotingRegressor
```

```
ensemble_model = [('br', linear_model.BayesianRidge()),
                  ('lr', linear_model.LinearRegression()),
                  ('rd', linear_model.Ridge()),
                 ('ls', linear_model.Lasso())]
```

```
vr = VotingRegressor(ensemble_model)
```

```
vr.fit(X_train, y_train)
vr_train_pred = vr.predict(X_train)
vr_test_pred = vr.predict(X_test)
```

```
# Evaluate model performance on train & test dataset.
vr_model_train_mae , vr_model_train_rmse, vr_model_train_r2 = evaluate_model(y_train, vr_train_pred)
vr_model_test_mae , vr_model_test_rmse, vr_model_test_r2 = evaluate_model(y_test, vr_test_pred)
```

```
print('Voting Regressor Model performance for Training set')
print("- Root Mean Squared Error: {:.4f}".format(vr_model_train_rmse))
print("- Mean Absolute Error: {:.4f}".format(vr_model_train_mae))
print("- R2 Score: {:.4f}".format(vr_model_train_r2))

print('----------------------------------')

print('Model performance for Test set')
print("- Root Mean Squared Error: {:.4f}".format(vr_model_test_rmse))
print("- Mean Absolute Error: {:.4f}".format(vr_model_test_mae))
print("- R2 Score: {:.4f}".format(vr_model_test_r2))
```

### Observation:
Ensemble of linear models gives approximately 92% accuracy which seems reasonable.
